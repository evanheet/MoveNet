```python
import torch
import torchvision
import torchvision.transforms as transforms

import os
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.autograd import Variable
import torch.utils.data as data
import torchvision
from torchvision import transforms

print("done")
```

```python
BATCH_SIZE = 4
TEST_DATA_PATH = "./images/test/"
TRANSFORM_IMG = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
    ])

test_data = torchvision.datasets.ImageFolder(root=TEST_DATA_PATH, transform=TRANSFORM_IMG)
test_data_loader  = torch.utils.data.DataLoader(test_data, batch_size=BATCH_SIZE, shuffle=True, num_workers=4)
print("Data loaded")
```

```python
# Build neural network and functions
class Net(torch.nn.Module):
    
    def __init__(self):
        super(Net, self).__init__()
        # Convoltion of 256*256 images with 3 RGB values, 6 5x5 filters (result is 256 - 5 + 1 = 252, or 252x252 pixels)
        self.conv1 = nn.Conv2d(3, 6, 5)
        
        # Pooling of 252*252 images (result is 252/2 = 126, or 126*126 pixels)
        self.pool = nn.MaxPool2d(2, 2)
        
        # Convoltion of 126*126 images, 16 5x5 filters (result is 126 - 5 + 1 = 122, or 122x122 pixels)
        # Pooled final time in "forward" function to obtain 16 61x61 images
        self.conv2 = nn.Conv2d(6, 16, 5)
        
        # Fully connected layers from 61*61*16 dimentions to 100, 100 to 60, and 60 to 2 (for 2 classes, person or not_person)
        self.fc1 = nn.Linear(61*61*16, 100)
        self.fc2 = nn.Linear(100, 60)
        self.fc3 = nn.Linear(60, 2)
        
    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = x.view(-1, 61*61*16)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x
            
MoveNet = Net()
print(MoveNet)
```

```python
# Load model parameters
MoveNet.load_state_dict(torch.load("Trained_Model.pth"))
print("Model parameters loaded")
```

```python
correct = 0
total = 0

# Print running list of successful predictions as testing data is run through model
# Print overall accuracy at the end
with torch.no_grad():
    for data in test_data_loader:
        X, y = data
        output = MoveNet(X)
        for idx, i in enumerate(output):
            if torch.argmax(i) == y[idx]:
                correct += 1
            total += 1
            print("Total tested: ", total, " ... Total correct: ", correct)
            
print("Overall accuracy: ", round(correct/total, 3))
```
